# Question 10

**Plot distribution of the match's duration vs walk distance one below the other.**

```
import matplotlib.pyplot as plt # importing matplot library
plt.style.use('classic') # using classic style
plt.figure() # intialising the figure
figure, axes = plt.subplots(nrows=2, ncols=1) # subplots with 1 row and 2 columns
# Subplot for match duration
plt.subplot(2,1,1) # plotting space at 1st row out of 2 rows
plt.title('Pubg players Match Duration plot')
plt.plot(df['matchDuration']) # plotting matchduration
plt.xlabel('Match Duration') # labeling x variable
plt.ylabel("Number of players") # labeling y varible
# subplot for walk distance
plt.subplot(2,1,2) #plotting space at 2nd row out of 2 rows
plt.title('Pubg players Match Duration plot')
plt.plot(df['walkDistance']) # plotting walkdistance
plt.xlabel("Walk Distance") # labeling x varible
plt.ylabel("Number of players") # labeling y varible
figure.tight_layout(pad=2.0) # giving a space between two plots by using padding parameter
plt.show() # showing the plots
```

# Question 22

**Take a sample of size 50 from the column damageDealt for 100 times and calculate its mean. Plot
it on a histogram and comment on its distribution.**

```
lst=[] # initialising empty list
for x in range(100):
  avg=np.mean(df['damageDealt'].sample(50)) # taking 50 samples from the dataset df
  lst.append(avg) # appending the mean value to the lst
sns.histplot(lst) # plotting histplot with the mean values
```
